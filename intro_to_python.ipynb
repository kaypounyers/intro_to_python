## <center> Common Methods </center>

## Methods are special operations used on variables. Methods will provide information about the variable or alter the variable and are VARIABLE TYPE specific. 
## Syntax: list.method(args)

```python
#LIST METHODS
```

```python
#Adding Elements to a List

list1 = ["you", "me", "us", "we"]                                          

list1.append("them")

print(list1)
```

```python
#Removing Elements from a List

list1 = ["you", "me", "us", "we"] 
list1.pop(2) #indexing - begins at 0
print(list1)
```

```python
#You can also remove a list element by referencing the element itself. 

list1 = ["you", "me", "us", "we"]                                                             
list1.remove("us")
print(list1)
```

```python
# Finding an Element's Index

list1 = ["you", "me", "us", "we"]                                           
meindex= list1.index("me")
print(meindex)
```

```python
#Counting Elements in a List

list1 = ["you", "me", "us", "we", "we"]                                                                     
wecount=list1.count("we")
print(wecount)
```

```python
#Join Lists

list1 = ["you", "me", "us", "we"]                   
list2 = ["Sheila", "Joe", "Ronald", "Jacob"]
list1.extend(list2)
print(list1)
```

```python
#Alphabetically Sort a List

list1 = ["you", "me", "us", "we"]                                                                  
list1.sort()
print(list1)
```

```python
#String Methods
```

```python
#Capitalizing the First Word of a String

string = "Tulane University Libraries"                        
string=string.capitalize()
print(string)
```

```python
#Capitalizing the First Letter of Each Word

string = "Tulane university libraries"                              
string = string.title()
print(string)
```

```python
#Making All Letters Lowercase

string = "Tulane University Libraries"                  
string = string.lower()
print(string)
```

```python
#Making All Letters Uppercase

string = "Tulane University Libraries"                               
string = string.upper()
print(string)
```

```python
#Dictionary Methods
```

```python
#Getting the Keys of a Dictionary

dct = {"shoe": 8, "shirt": "medium", "wpants": 34, "lpants": 32}         
dct.keys()   
```

```python
#Getting the Values of a Dictionary

dct = {"shoe": 8, "shirt": "medium", "wpants": 34, "lpants": 32}     
dct.values()
```

```python
#Getting the Key-Value Pairs of a Dictionary

dct = {"shoe": 8, "shirt": "medium", "wpants": 34, "lpants": 32}     
dct.items()    
```

```python
#Getting a Value by Providing a Key

dct = {"shoe": 8, "shirt": "medium", "wpants": 34, "lpants": 32}     
dct.get("shoe")   
```

```python
#Delete a Key-Value Pair from a Dictionary

dct = {"shoe": 8, "shirt": "medium", "wpants": 34, "lpants": 32} 
dct.pop("shoe")
print(dct) 
```

```python
#Add a Key-Value Pair to a Dictionary

dct = {"shoe": 8, "shirt": "medium", "wpants": 34, "lpants": 32} 

dct.update({"hat":"small"})
print(dct)
```
